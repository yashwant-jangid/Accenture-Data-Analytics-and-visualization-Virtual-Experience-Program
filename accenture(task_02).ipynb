DATA CLEANING AND MODELLING

```python
import pandas as pd
import numpy as np
import matplotlib as plot
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
path = "/content/drive/MyDrive/Content.csv"
content_df = pd.read_csv(f"{path}")
```

```python
path = "/content/drive/MyDrive/Reactions.csv"
reactions_df = pd.read_csv(f"{path}")
path = "/content/drive/MyDrive/ReactionTypes.csv"
types_df = pd.read_csv(f"{path}")
```

```python
content_df.drop('Unnamed: 0',axis=1,inplace=True)
content_df.head()
```

```python
reactions_df.drop(['Unnamed: 0','User ID'],axis=1,inplace=True)
reactions_df.head()
```

```python
types_df.drop(['Unnamed: 0'],axis=1,inplace=True)
types_df.head()
```

```python
merged_df = content_df.merge(reactions_df, on=['Content ID'], how='left')
merged_df.head()
```

```python
merged_df = merged_df.rename(columns={'Type_x':'Cont_Type','Type_y':'Type'})
merged_df = merged_df.merge(types_df, on=['Type'], how='left')
merged_df.head()
```

```python
merged_df.isnull().sum()
```

```python
merged_df.drop(['URL'],axis=1,inplace=True)
```

```python
merged_df=merged_df.dropna()
```

```python
merged_df.isnull().sum()
```

```python
merged_df
```

```python
merged_df['Category'].unique()
```

```python
merged_df['Category']= merged_df['Category'].replace('"', '', regex=True)
merged_df['Category']=merged_df['Category'].str.lower()
merged_df['Category'].unique()
```

```python
merged_df.groupby('Category').agg({"Type" : "count"}).reset_index().sort_values(by='Type', ascending=False).head(5)
```

```python
merged_df.groupby('Category').agg({"Score" : "sum"}).reset_index().sort_values(by='Score', ascending =False).head()
```

```python
from google.colab import files
merged_df.to_csv('data.csv', encoding = 'utf-8-sig')
files.download('data.csv')
```
