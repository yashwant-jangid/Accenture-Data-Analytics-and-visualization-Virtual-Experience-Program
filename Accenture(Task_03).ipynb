```python
path = "/content/drive/MyDrive/data.csv"
data = pd.read_csv(f"{path}")
```

```python
data.head()
```

```python
data['Datetime'] = pd.to_datetime(data['Datetime'] ,infer_datetime_format=True)
```

```python
data.Datetime
```

```python
data["Month"] = data["Datetime"].dt.month
```

```python
data.head()
```

```python
data["Count"] = data.groupby('Month')['Month'].transform('count')
```

```python
data.head()
```

```python
data_sorted = data.sort_values(by='Count', ascending=False)

highest_count = data_sorted['Count'].iloc[0]
highest_month = data_sorted['Month'].iloc[0]

plt.figure(figsize=(10, 6))

bars = plt.bar(data['Month'], data['Count'], color=['blue' if month != highest_month else 'red' for month in data['Month']])

plt.xlabel('Month')
plt.ylabel('Count')
plt.title('Count by Month')
```
